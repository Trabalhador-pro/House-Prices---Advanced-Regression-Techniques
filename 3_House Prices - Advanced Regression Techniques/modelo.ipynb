## Transformação

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler


colunas_num = X.select_dtypes(include=['int64', 'float64']).columns.to_list()
colunas_cat = X.select_dtypes(include=['object']).columns.to_list()

Pipeline_num = Pipeline([
    ('input', SimpleImputer(strategy='median')),
    ('escalonamento', StandardScaler())
    
])

Pipeline_cat = Pipeline([
    ('input', SimpleImputer(strategy='constant', fill_value='0')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

processamento = ColumnTransformer([
    ('num', Pipeline_num, colunas_num),
    ('cat', Pipeline_cat, colunas_cat)
])

Pipeline_final = Pipeline([
    ('Processamento', processamento)
])

X_train = Pipeline_final.fit_transform(X_train)

X_test = Pipeline_final.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge

modelo_1 = LinearRegression()

modelo_1.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

def teste_modelo (modelo, x_treino, x_teste, y_treino, y_teste ):

    y_predict_train = modelo.predict(x_treino)

    y_predict_test = modelo.predict(x_teste)

    mbe_train = mean_absolute_error(y_treino, y_predict_train)

    mbe_test = mean_absolute_error(y_teste, y_predict_test)

    print(f'O Erro quadratico do treino foi de {mbe_train:.2f}')
    print(f'O Erro quadratico do teste foi de {mbe_test:.2f}')

teste_modelo(modelo_1, X_train, X_test, y_train, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

modelo_2 = RandomForestRegressor(n_estimators=200,
                                 criterion='squared_error',
                                 min_samples_leaf=4,
                                 random_state=42,
                                 )

modelo_2.fit(X_train, y_train)

teste_modelo(modelo_2, X_train, X_test, y_train, y_test)
```

```python
from xgboost import XGBRegressor

modelo_3 = XGBRegressor(n_estimators=1000,
                        learning_rate=0.04,
                        max_depth=4,
                        eval_metric="rmse",
                        random_state=42)

modelo_3.fit(X_train, y_train)

teste_modelo(modelo_3, X_train, X_test, y_train, y_test)
```

```python
from sklearn.ensemble import StackingRegressor

modelo_stacking = StackingRegressor(
    estimators=[('linear', modelo_1), ('arvore', modelo_2), ('bossting',modelo_3)],
    final_estimator=Ridge(),
    cv=5,
    passthrough=False
)

modelo_stacking.fit(X_train, y_train)

teste_modelo(modelo_stacking, X_train, X_test, y_train, y_test)
```

```python
df_teste_final

df_teste_final_tratado = df_teste_final.drop('Id', axis=1)

df_teste_final_tratado = Pipeline_final.transform(df_teste_final_tratado)

predict = modelo_stacking.predict(df_teste_final_tratado)

teste = pd.DataFrame({
    'id': df_teste_final['Id'],
    'SalePrice': predict
})


teste.to_csv('submission.csv', index=False)
```
